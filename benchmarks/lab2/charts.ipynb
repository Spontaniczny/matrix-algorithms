```julia
import matplotlib.pyplot as plt
import pandas as pd
```

```julia
# column_types_times = {'column_name1': int, 'column_name2': int}
# func 1 -> LU
# func 2 -> Det
# func 3 -> Inv
with open("operations.csv", 'r') as file:
    operations = pd.read_csv(file, dtype = int)
    
with open("times.csv", 'r') as file:
    times = pd.read_csv(file, dtype = float)
```

```julia
operations
```

```julia
times
```

```julia
lu_data = operations[operations["function"] == 1]
det_data = operations[operations["function"] == 2]
inv_data = operations[operations["function"] == 3]

lu_time = times[times["function"] == 1]
det_time = times[times["function"] == 2]
inv_time = times[times["function"] == 3]

lu_time["time"] = 100 * lu_time["time"]
det_time["time"] = 100 * det_time["time"]
inv_time["time"] = 100 * inv_time["time"]
```

```julia
lu_data["total_operations"] = lu_data["add"] + lu_data["mul"]
det_data["total_operations"] = det_data["add"] + det_data["mul"]
inv_data["total_operations"] = inv_data["add"] + inv_data["mul"]
```

```julia
lu_data
```

```julia
lu_time
```

```julia
plt.scatter(lu_time["size"], lu_time["time"])
# plt.legend()
plt.xticks(list(2**i for i in range(5, 10)))
plt.title("Time of LU")
plt.xlabel("Matrix size")
plt.ylabel("Time")
plt.show()
```

```julia
plt.scatter(lu_data["size"], lu_data["add"])
# plt.legend()
plt.xticks(list(2**i for i in range(5, 10)))
plt.title("Number of additions in LU")
plt.xlabel("Matrix size")
plt.ylabel("Number of additions")
plt.show()
```

```julia
plt.scatter(lu_data["size"], lu_data["mul"])
# plt.legend()
plt.xticks(list(2**i for i in range(5, 10)))
plt.title("Number of multiplications in LU")
plt.xlabel("Matrix size")
plt.ylabel("Number of multiplications")
plt.show()
```

```julia
plt.scatter(lu_data["size"], lu_data["total_operations"])
# plt.legend()
plt.xticks(list(2**i for i in range(5, 10)))
plt.title("Number of operations in LU")
plt.xlabel("Matrix size")
plt.ylabel("Number of operations")
plt.show()
```

```julia
plt.scatter(det_time["size"], det_time["time"])
# plt.legend()
plt.xticks(list(2**i for i in range(5, 10)))
plt.title("Time of determinant")
plt.xlabel("Matrix size")
plt.ylabel("Time")
plt.show()
```

```julia
plt.scatter(det_data["size"], det_data["add"])
# plt.legend()
plt.xticks(list(2**i for i in range(5, 10)))
plt.title("Number of additions in det")
plt.xlabel("Matrix size")
plt.ylabel("Number of additions")
plt.show()
```
