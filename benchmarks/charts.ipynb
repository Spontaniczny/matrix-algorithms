```julia
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```julia
# column_types_times = {'column_name1': int, 'column_name2': int}

with open("binet/times.csv", 'r') as file:
    csv_binet_times = pd.read_csv(file, dtype = int)

with open("binet/operations.csv", 'r') as file:
    csv_binet_operations = pd.read_csv(file, dtype = int)

with open("strassen/times.csv", 'r') as file:
    csv_strassen_times = pd.read_csv(file, dtype = int)
    
with open("strassen/operations.csv", 'r') as file:
    csv_strassen_operations = pd.read_csv(file, dtype = int)
```

```julia
plt.scatter(csv_binet_times["size"], csv_binet_times["time"], label = "Binet")
plt.scatter(csv_strassen_times["size"], csv_strassen_times["time"], label = "Strassen")
# plt.scatter AI later
plt.legend()
plt.xticks(list(2**i for i in range(4, 9)))
plt.yticks(np.linspace(min(csv_binet_times["time"]), max(csv_binet_times["time"]), num = 10))
plt.title("Comparison of times")
plt.xlabel("Matrix size")
plt.ylabel("Time in milliseconds")
plt.show()
```

```julia
plt.scatter(csv_binet_operations["size"], csv_binet_operations["additions"], label = "Binet")
plt.scatter(csv_strassen_operations["size"], csv_strassen_operations["additions"], label = "Strassen")
# plt.scatter AI later
plt.legend()
plt.xticks(list(2**i for i in range(4, 9)))
plt.yticks(np.linspace(min(csv_strassen_operations["additions"]), max(csv_strassen_operations["additions"]), num = 10))
plt.title("Comparison of numbers of additions")
plt.xlabel("Matrix size")
plt.ylabel("Number of additions")
plt.show()
```

```julia
plt.scatter(csv_binet_operations["size"], csv_binet_operations["multiplications"], label = "Binet")
plt.scatter(csv_strassen_operations["size"], csv_strassen_operations["multiplications"], label = "Strassen")
# plt.scatter AI later
plt.legend()
plt.xticks(list(2**i for i in range(4, 9)))
plt.yticks(np.linspace(min(csv_binet_operations["multiplications"]), max(csv_binet_operations["multiplications"]), num = 10))
plt.title("Comparison of numbers of multiplications")
plt.xlabel("Matrix size")
plt.ylabel("Number of multiplications")
plt.show()
```
